Exploratory Data Analysis:
- In this section we will analyze profit,categories,regoins,customers and product

### EDA1 which category has the highest total sales:
-I group the category to conclude the total sales of the category

```python
df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
```

### after grouping the category with sales we can clearly see that the highest slaes are Technology

### EDA 2 grouping top 10 subcategories with sales

```python
df.groupby('Sub-Category')['Sales'].sum().sort_values(ascending=False).head(10)
```

#### atfer observation here is the given top 10 subcategories

### EDA3 Which region is the most profitable

```python
df.groupby('Region')['Profit'].sum().sort_values(ascending=False)
```

#### The most Profitable region is West

### EDA4 Which segment is the most profitable 

```python
df.groupby('Segment')['Profit'].sum().sort_values(ascending=False)
```

#### consumer segment is the most profitable

### EDA5 Top 10 product by total profit

```python
df.groupby('Product Name')['Profit'].sum().sort_values(ascending=False)
```

#### Here are the given top 10 profitable product

### EDA6 Sub Categories which are in loss

```python
df.groupby('Sub-Category')['Profit'].sum().sort_values()
```

### EDA7 Top 10 customers by total sales

```python
df.groupby('Customer Name')['Sales'].sum().sort_values(ascending=False).head(10)
```

#### Here are the top 10 customer

### EDA8 Which Ship mode is most frequently used

```python
df['Ship Mode'].value_counts()
```

#### These are the most frequently ship mode used

### EDA9 Total sales by region and category

```python
df.pivot_table(values='Sales',index='Region',columns='Category',aggfunc='sum')
```

#### by this data we can see which region performs well and in which category

### Date Based Analysis

```python
df['Order Date']=pd.to_datetime(df['Order Date'])
df['Ship Date']=pd.to_datetime(df['Ship Date'])
```

#### in the given information of the datatypes of the column the datatype of Order Date and Ship Date is in object type so i first converted it in datetime

### EDA10 Sales per Year

```python
df['Year']=df['Order Date'].dt.year
df.groupby('Year')['Sales'].sum().sort_index()
```

#### As we can every year sales have been increasing we can clearly see the highest sale is in the year 2017

### EDA10 Sales per month

```python
df['Month']=df['Order Date'].dt.month
df.groupby('Month')['Sales'].sum().sort_index()
```
